# Obtención de datos sobre radiación solar y visualización de los mismos. 

[Solar_Radiation_Prediction (Random Forest)](https://www.kaggle.com/code/rachelswati/solar-radiation-prediction-random-forest/notebook)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python


data=pd.read_csv('SolarPrediction.csv')

```

```python


data.head()

```

# Data Exploration and Data Visualization

```python


data.info()

```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python


plt.figure(figsize=(12,6))
sns.heatmap(data.corr(),annot=True)

```

```python
sns.histplot(x=data['Temperature'],bins=20)

plt.show() 
```

```python
sns.histplot(x=data['Pressure'],bins=20)

plt.show() 
```

```python
plt.figure(figsize=(10,6),dpi=150)
sns.scatterplot(x=data['Temperature'],y=data['Pressure'],hue=data['Radiation'])


plt.savefig('temperatura_presion_radiacion.jpg')
plt.show()
```

```python


sns.histplot(x=data['WindDirection(Degrees)'],bins=20)

```
